```python
import pandas as pd
```

## Explore IMDb Dataset

```python
movies = pd.read_csv('data/imdb_movies.csv')
movies.head()
```

```python
movies.info()
```

```python
movies.drop(columns=['budget', 'usa_gross_income', 'worlwide_gross_income', 'metascore'], inplace=True)
movies.head()
```

```python
names = pd.read_csv('data/imdb_names.csv')
names.head()
```

```python
names.info()
```

```python
names.drop(columns=['height', 'bio', 'birth_details', 'date_of_birth', 'place_of_birth', 'death_details', 'date_of_death', \
                   'place_of_death', 'reason_of_death', 'spouses_string'], inplace=True)
names.head()
```

```python
ratings = pd.read_csv('data/imdb_ratings.csv')
ratings.head()
```

```python
ratings.info()
```

```python
ratings.drop(columns=['allgenders_0age_avg_vote', 'allgenders_0age_votes', 'males_0age_avg_vote', 'males_0age_votes', \
                      'females_0age_avg_vote', 'females_0age_votes'], inplace=True)
ratings.head()
```

```python
principals = pd.read_csv('data/imdb_title_principals.csv')
principals.head()
```

```python
principals.info()
```

```python
principals.drop(columns=['job', 'characters'], inplace=True)
principals.head()
```

```python
full_df = movies.merge(ratings, how='left', on='imdb_title_id')\
                .merge(principals, how='left', on='imdb_title_id')\
                .merge(names, how='left', on='imdb_name_id')
full_df.head()                
```

```python
full_df.info()
```

```python
full_df.to_csv('data/clean_data.csv')
```

## Explore MovieLens Data Set
